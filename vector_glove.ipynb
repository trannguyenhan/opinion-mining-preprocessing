```python
import pandas as pd
import numpy as np
```

```python
dictionary = open('dictionary.txt', 'r').read().split('\n')
dictionary
```

```python
glove_file = open('glove.6B.100d.txt', 'r')
```

```python
glove_dictionary = open('glove_dictionary_vector.txt', 'w')
```

```python
len(dictionary)
```

```python
cnt = 0
while True:
  line = glove_file.readline()

  if not line: 
    break
  
  word = line.split(" ")[0]
  if word in dictionary:
    cnt += 1
    
    if cnt > len(dictionary):
      break
    
    glove_dictionary.write(line)

glove_dictionary.close()
glove_file.close()
```
